### Education Level

```python
data["What is the highest level of education you have completed?"].value_counts()
```

```python
# Function to reduce the length of unique values
def reduce_length(value):
    if value.startswith('Postgraduate Degree'):
        return 'Postgraduate Degree'
    else:
        return value

# Apply the function to the 'Education Level' column
data['What is the highest level of education you have completed?'] = data['What is the highest level of education you have completed?'].apply(reduce_length)
```

```python

# Suppose 'Education Level' is the column name containing education levels
education_counts = data['What is the highest level of education you have completed?'].value_counts()

# Calculate percentages
total = education_counts.sum()
percentages = (education_counts / total) * 100

# Plotting the bar graph
plt.figure(figsize=(10, 8))
bars = plt.bar(education_counts.index, education_counts.values)

#plt.bar(education_counts.index, education_counts.values, color='skyblue')

# Adding labels and title
plt.xlabel('Education Level')
plt.ylabel('Percentage Count')
plt.title('Highest Level of Education Completed')

# Rotating x-axis labels for better readability
plt.xticks(rotation=360)

# Annotating bars with counts and percentages
for bar, count, percent in zip(bars, education_counts.values, percentages):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'({percent:.1f}%)',
             ha='center', va='bottom', fontsize=9)

# Displaying the bar graph
plt.tight_layout()
plt.show()
```

### Mode of Payment of HealthCare Services

```python
data["How do you pay for the majority of your medical expenses?"].value_counts()
```

```python

# Suppose 'Education Level' is the column name containing education levels
payment_counts = data['How do you pay for the majority of your medical expenses?'].value_counts()

# Sort the data by counts in descending order
payment_counts = payment_counts.sort_values(ascending=True)

# Calculate percentages
total = payment_counts.sum()
percentages = (payment_counts / total) * 100

# Plotting the horizontal bar graph
plt.figure(figsize=(14, 6))
bars = plt.barh(payment_counts.index, payment_counts.values)

# Adding labels and title
plt.xlabel('Count')
plt.ylabel('Mode of Payment')
plt.title('Mode of Payment of HealthCare Services')

# Annotating bars with percentages only
for bar, percent in zip(bars, percentages):
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height() / 2, f'{percent:.1f}%',
             ha='left', va='center', fontsize=9)

# Displaying the bar graph
plt.tight_layout()
plt.show()
```
